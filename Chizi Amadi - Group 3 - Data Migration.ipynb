```python
# Create connection
conn = pymysql.connect(
    host='localhost',
    port=3306,
    user='root',
    password='9175',
    db='database1'
)
```

```python
# Create cursor
cursor = conn.cursor()
```

```python
# Load  customer dataframe into database
for index, row in customer.iterrows():
    sql = 'INSERT INTO customer (customer_id, age, gender, geography) VALUES (%s, %s, %s, %s)'
    val = (int(row['customer_id']), row['age'], row['gender'], row['geography'])
    cursor.execute(sql, val)
```

```python
# Commit changes
conn.commit()
```

```python
# Load  comapny dataframe into database
for index, row in company.iterrows():
    sql = 'INSERT INTO company (company_id, estimated_salary, tenure, customer_id) VALUES (%s, %s, %s, %s)'
    val = (int(row['company_id']), row['estimated_salary'], row['tenure'], row['customer_id'])
    cursor.execute(sql, val)
```

```python
# Commit changes
conn.commit()
```

```python
# Load  credit dataframe into database
for index, row in credit.iterrows():
    sql = 'INSERT INTO credit (company_id, credit_id, customer_id, credit_score, balance, num_of_products, has_cr_card, is_active_member) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)'
    val = (int(row['company_id']), row['credit_id'], row['customer_id'], row['credit_score'], row['balance'], row['num_of_products'], row['has_cr_card'], row['is_active_member'])
    cursor.execute(sql, val)
```

```python
# Commit changes
conn.commit()
```

```python
# Close connection
conn.close()
```
